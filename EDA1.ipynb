```python
# Fresh load
movies = pd.read_csv('/content/drive/MyDrive/Movies.csv')

# Scale budget and Gross
movies['budget'] = movies['budget'] / 1000000
movies['Gross'] = movies['Gross'] / 1000000

# Step 1 - Convert MetaCritic to scale of 10
movies['MetaCritic'] = movies['MetaCritic'] / 10

# Step 2 - Create Avg_rating
movies['Avg_rating'] = (movies['MetaCritic'] + movies['IMDb_rating']) / 2

# Step 3 - Sort descending
movies = movies.sort_values('Avg_rating', ascending=False)

# Step 4 - UniversalAcclaim
UniversalAcclaim = movies[
    (abs(movies['IMDb_rating'] - movies['MetaCritic']) < 0.5) &
    (movies['Avg_rating'] >= 8)
]
UniversalAcclaim
```

```python
df_by_genre = movies.copy()
```

```python
df_by_genre['cnt'] = 1
```

```python
genre1 = df_by_genre.groupby('genre_1').sum(numeric_only=True)
genre2 = df_by_genre.groupby('genre_2').sum(numeric_only=True)
genre3 = df_by_genre.groupby('genre_3').sum(numeric_only=True)
```

```python
df_by_genre = genre1.add(genre2, fill_value=0).add(genre3, fill_value=0)
```

```python
df_by_genre = df_by_genre[df_by_genre['cnt'] >= 10]
```

```python
df_by_genre = df_by_genre.div(df_by_genre['cnt'], axis=0)
```

```python
df_by_genre = df_by_genre.div(df_by_genre['cnt'], axis=0)
```

```python
vote_cols = [col for col in df_by_genre.columns if 'Votes' in col or 'CVotes' in col]
df_by_genre[vote_cols] = df_by_genre[vote_cols].round(2)
df_by_genre
```

```python
cvotes_cols = [col for col in df_by_genre.columns if col.startswith('CVotes')]
df_by_genre[cvotes_cols] = df_by_genre[cvotes_cols].astype(int)
df_by_genre.head()
```

```python
genre_top10 = df_by_genre.copy()

sns.barplot(x=genre_top10.index, y=genre_top10['cnt'])
plt.xticks(rotation=45)
plt.title('Genre Counts')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
male_cols = ['CVotesU18M','CVotes1829M','CVotes3044M','CVotes45AM']
female_cols = ['CVotesU18F','CVotes1829F','CVotes3044F','CVotes45AF']

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.heatmap(genre_top10[male_cols], annot=True, fmt='g', ax=axes[0])
axes[0].set_title('Male CVotes by Genre')

sns.heatmap(genre_top10[female_cols], annot=True, fmt='g', ax=axes[1])
axes[1].set_title('Female CVotes by Genre')

plt.tight_layout()
plt.show()
```

```python
male_vote_cols = ['VotesU18M','Votes1829M','Votes3044M','Votes45AM']
female_vote_cols = ['VotesU18F','Votes1829F','Votes3044F','Votes45AF']

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.heatmap(genre_top10[male_vote_cols], annot=True, fmt='.2f', ax=axes[0])
axes[0].set_title('Male Ratings by Genre')

sns.heatmap(genre_top10[female_vote_cols], annot=True, fmt='.2f', ax=axes[1])
axes[1].set_title('Female Ratings by Genre')

plt.tight_layout()
plt.show()
```

```python
# Reload clean movies first since it got filtered earlier
movies = pd.read_csv('/content/drive/MyDrive/Movies.csv')
movies['budget'] = movies['budget'] / 1000000
movies['Gross'] = movies['Gross'] / 1000000

# Create IFUS column
movies['IFUS'] = movies['Country'].apply(lambda x: 'USA' if x == 'USA' else 'non-USA')

# Boxplot 1: CVotes (number of votes)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(x='IFUS', y='CVotesUS', data=movies, ax=axes[0])
axes[0].set_title('US Votes: US vs non-US Movies')
sns.boxplot(x='IFUS', y='CVotesnUS', data=movies, ax=axes[1])
axes[1].set_title('Non-US Votes: US vs non-US Movies')
plt.show()

# Boxplot 2: Ratings
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(x='IFUS', y='VotesUS', data=movies, ax=axes[0])
axes[0].set_title('US Ratings: US vs non-US Movies')
sns.boxplot(x='IFUS', y='VotesnUS', data=movies, ax=axes[1])
axes[1].set_title('Non-US Ratings: US vs non-US Movies')
plt.show()
```
