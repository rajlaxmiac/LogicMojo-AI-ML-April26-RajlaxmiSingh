```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BikeIndia.csv')

df.head()

```

```python
df[['dteday', 'cnt']]
```

```python


# Example 1 — Get all columns for row 0
df.loc[0]

# Example 2 — Get only date and cnt column
df.loc[0:5, ['dteday', 'cnt']]

# Example 3 — Get rows where season is Summer (3)
df.loc[df['season'] == 3, ['dteday', 'cnt']]
```

```python
df[df['cnt'] > 4000][['dteday', 'cnt']]
```

```python
# Rainy days (weathersit=3) AND still high rentals
df[(df['weathersit'] == 3) & (df['cnt'] > 1000)][['dteday','weathersit','cnt']]

```

```python
# Summer OR Fall high rental days
df[(df['season'] == 3) | (df['season'] == 4)][['dteday','season','cnt']]
```
