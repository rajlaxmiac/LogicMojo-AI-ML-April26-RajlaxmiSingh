```python
# my list with duplicates
my_list = [8, 9, 2, 2, 3, 4, 5, 2]

new_list = []


for x in my_list:
    if x not in new_list:
        new_list.append(x)


print(new_list)
```

```python
# given dictionary
my_dict = {'Mobile': ['Redmi', 'Samsung', 'Realme'],
           'Laptop': ['Dell', 'HP'],
           'TV': ['Videocon', 'Sony']}


result = []

# loop through each key in dictionary
for key in my_dict:
    # loop through each value in the list of that key
    for value in my_dict[key]:
        new_string = key + "_" + value
        result.append(new_string)

# print final list
print(result)
```
